```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([0,7,10])
ypoints=np.array([0,26,33])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=([0,25])
ypoints=([0,6])
plt.plot(xpoints,ypoints, "3")
plt.show()
```

```python
ypoints=np.array([1,5,8,10,6])
plt.plot(ypoints, marker="d")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mec = 'g', mfc='y')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed', color = 'black', linewidth='3')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
y3 = ([5, 3, 4, 8])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])


plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("Plot Title")

plt.plot(x, y)
plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.xlabel("SOMETHING")
plt.ylabel("SOMEONE")
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color="black")


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='red')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y,color="black",s=sizes, alpha=0.2)

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
